```python
import requests
res=requests.get ("https://www.google.com/search?q=%E4%BC%8A%E8%8E%89%E8%8E%8E%E7%99%BD&rlz=1C5CHFA_enTW1034TW1034&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjloOWauvr7AhUQCaYKHS6TBMQQ_AUoAXoECAEQAw&biw=1440&bih=821&dpr=2")

from bs4 import BeautifulSoup
soup=BeautifulSoup(res.text)
n=0
import shutil  #將資料下載


#print(soup.prettify())
#print((soup("img",class_="yWs4tf")[0])["src"])

#img=soup("img",class_="yWs4tf")
#print(type(img[0]))


for img in soup("img",class_="yWs4tf") :
    
    pic = requests.get(img["src"] , stream=True)   #request 圖片網址，須先查看適用get還是post
                                                                    #stream=T,因圖片檔案較大，需增加流量才能取得

    
    n+=1
    fname="/Users/xuan/python/imgdownload/伊莉莎白"+str(n)+".jpg"
    
    with open(fname,"wb") as f:             # wb 以二進位方式寫入檔案 #whith 可於讀取完直接close
        
            shutil.copyfileobj(pic.raw,f)      #(要讀取的檔案,要放入的位置)
            
print("done")
    
```

```python
import requests
url="https://goodinfo.tw/tw/StockIdxHisAna.asp?STOCK_ID=%E6%AB%83%E8%B2%B7%E6%8C%87%E6%95%B8"


#破解阻擋爬蟲設定
headers = {"user-agent":"Chrome/108.0.0.0" ,"content-type":"text/html"}
data=requests.get(url,headers=headers) 


#將uts-16(亂碼)轉成uts-8
data.encoding = "utf-8" 


from bs4 import BeautifulSoup
soup2=BeautifulSoup(data.text)

print(soup2.prettify())
```
