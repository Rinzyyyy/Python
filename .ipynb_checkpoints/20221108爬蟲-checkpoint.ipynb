```python
res="""

<html>
  <body>
         <a href="https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A" >這是一行字</a>
         <a> 第2行 </a>
         <a> 第3行 </a>
         <h1 class="abc" href="123">abc</h1>
         <a> 第4行 </a>
         <a> 第5行 </a>
          <h1 class="job">職缺：<h2>大數據工程師</h2> </h1>
      </body>
</html>

"""
#html 標籤--屬性名=>屬性內容--標籤值
# tag-name--attr_name=>attr_valu--tag_value


from bs4 import BeautifulSoup #翻譯html給python

soup=BeautifulSoup(res)

print(soup)
print("-----------------------------------------")
print("")

print(soup.h1) #可直接跳級搜尋-建議當tag名稱獨一無二時使用否則只會找到第一個
print("")

print("-----sting&text-------")
print("回傳純字串內容＆跨階層回傳字串內容")
print("")
print(soup.body.h1.string) #可依續搜尋
print(soup.body.h1.text)
print(soup.body.a.text)
print(soup.body.select('h1')[1].text)  # h1內含h2非單純字串內容 text全輸出
print(soup.body.select('h1')[1].string) # h1內含h2非單純字串內容 string拒絕處理
print(soup.body.select('h1')[1].text.replace(soup.select('h1')[1].h2.string," "))  # text找到全部再用str找到多餘的部分並用空值取代
print(soup.select('h1')[1].h2.string)
print("")

print("-----select-------")
print("搜索回傳list")
print("")
print(soup.body.select("h1")) # type為list
print(soup.body.select("h1")[1])
print("")

print("-----find_all&find-------")
print("#雙重條件搜索回傳list")
print("")
print(soup.find_all("h1",class_="job")[0]) #雙重條件搜索回傳list
print(soup.find("h1",class_="job")) 
print("")

print("-----[attr_name]-------")
print("回傳attr_value")
print("")
print(soup.a["href"]) #[tag] 顯示tag的內容（屬性）

```
