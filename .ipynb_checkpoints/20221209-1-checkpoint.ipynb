```python
#提供輸入8個數字，代表四個點（x,y)（在平面上不重複）,由點判斷是正方形，長方形，菱形，還是其他

x1,y1,x2,y2,x3,y3,x4,y4=map(int,input("輸入數字").split())

"""D1=[]
D2=[]
D3=[]
D4=[]

D1.extend([x1,y1])
D2.extend([x2,y2])
D3.extend([x3,y3])
D4.extend([x4,y4])"""


D1=[x1,y1]
D2=[x2,y2]
D3=[x3,y3]
D4=[x4,y4]

dis1=((x1 - x2)**2 + (y1 - y2)**2)**1/2
dis2=((x2 - x4)**2 + (y2 - y4)**2)**1/2
dis3=((x3 - x4)**2 + (y3 - y4)**2)**1/2
dis4=((x1 - x3)**2 + (y1 - y3)**2)**1/2
dis5=((x1 - x4)**2 + (y1 - y4)**2)**1/2
dis6=((x3 - x2)**2 + (y3 - y2)**2)**1/2

print(D1)
print(D2)
print(D3)
print(D4)

line = [dis1,dis2,dis3,dis4,dis5,dis6]
print(line)


if dis1==dis2==dis4==dis3 and dis5==dis6 :
     print("正方形")
    
elif dis1==dis3 and dis4==dis2 and dis1!=dis4 :
    print("長方形")
        
elif dis1==dis3==dis4==dis2 and dis5!=dis6 : 
    print("菱形") 

else :
    print("其他")





```

```python
"""import random
#place = [random.randint(1,10)for i in range(8)]
#place = [0,0,1,1,0,1,1,0]
#place = [0,0,2,1,0,1,2,0]
place = [2,0,0,1,0,-1,-2,0]
#place = [1,0,0,10,1,20,2,10]"""

place=[]
place.extend(map(int,input("輸入數字").split()))
print(place)

print("["+str(place[0])+","+str(place[1])+"]")#點座標 
print("["+str(place[2])+","+str(place[3])+"]") #字串才能加字串，用加號才不會像逗號會有空格
print("["+str(place[4])+","+str(place[5])+"]")
print("["+str(place[6])+","+str(place[7])+"]")

a = (((place[0]-place[2])**2)+((place[1]-place[3])**2))**0.5 #連線長度
b = (((place[0]-place[4])**2)+((place[1]-place[5])**2))**0.5
c = (((place[0]-place[6])**2)+((place[1]-place[7])**2))**0.5
d = (((place[2]-place[4])**2)+((place[3]-place[5])**2))**0.5
e = (((place[2]-place[6])**2)+((place[3]-place[7])**2))**0.5
f = (((place[4]-place[6])**2)+((place[5]-place[7])**2))**0.5

line = sorted([a,b,c,d,e,f]) #依照小到大排序，因為a~f沒有特別設定是哪裡的距離

print(line)

if line[0] == line[1] == line[2] == line[3] and line[4]==line[5]:
    print("正方形")
elif line[0] == line[1] == line[2] == line[3] or line[4] == line[1] == line[2] == line[3] and line[4]!=line[5]:
    print("菱形")
elif line[0] == line[1] and line[2] == line[3] and line[4]==line[5]:
    print("長方形")
else:
    print("不規則四邊形")
```
