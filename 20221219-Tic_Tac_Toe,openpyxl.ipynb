```python
import pandas as pd
print("---------用list或Dic創建表格----------")
g={"kj":["q",3,"e",4],"rk":[1,2,3,4]}
df=pd.DataFrame(g)
print(df)

print("-------------")

g=[["kj","q",3,"e",4],["rk",1,2,3,4]]
dfe=pd.DataFrame(g)
print(dfe)

print("---------更改欄位名稱----------")
dfe.columns=["a","b","d","e","c"]
dfe.index=["s1","s2"]
print(dfe)

print("---------取得前/後(head/tail)幾筆數值--------")
two=df.head(2) 
print(two)

print("--------搜尋/修改-at[列,欄名]/iat[列,欄]＝修改值--------")
print(df.iat[1,0])
df.iat[1,0]=9
print(df.iat[1,0])

print("--------loc[[列,列],[欄名,欄名]]/iloc[[列,列],[欄,欄]]--------")
print(df.loc[[1,0],["kj"]])

print("--------新增欄資料 insert(插入位置,colomn=欄名稱,value=[值])--------")
df.insert(2,column="hh",value=[2,3,4,5])
print(df)

print("--------合併資料 concat([data1,data2],ignore_index=True)--------")
df2=pd.DataFrame({"kj":[3],"rk":[3],"hh":[3]})
df3=pd.concat([df,df2],ignore_index=True) #ignore可省略，True表示重新排序
print(df3)

print("--------刪除資料 data名.drop([欄名/列名],axis=0/1 (刪除列/欄))--------")
#drop＿duplicates刪除重複
#dropna刪除空值

df3.drop([4,3],axis=0)

print("--------篩選資料 --------")
print(df[df["kj"].isin(["e"])])

df5=pd.read_excel("/Users/xuan/python/1219xlsx")
#df=pd.to_excel("/Users/xuan/python/1219xlsx")
```
