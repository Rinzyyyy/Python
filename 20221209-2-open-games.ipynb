```python
import random
# 首先做一個6*6大小的迷宮
#列印出所走的路徑以及地圖，路徑以3表示
maze = [[2, 2, 2, 2, 2, 2, 2],
        [2, 0, 0, 0, 0, 0, 2],
        [2, 0, 2, 0, 2, 0, 2],
        [2, 0, 0, 2, 0, 0, 2],
        [2, 2, 0, 2, 0, 2, 2],
        [2, 1, 0, 2, 0, 4, 2],
        [2, 2, 2, 2, 2, 2, 2]]
#向上走的函數
def goUp(location):
#如果位置已在最上面那排，則不需要向上走
    if(location[0]== 0):
         return False
    else:
#如果不是在最上排，則向上走
        newLocation = [location[0]-1, location[1]]
#如果已經走過，就回傳False
        if newLocation in routeHistory:
            routeHistory.append(newLocation)
            return False
#如果這個位置是牆，則回傳False
        elif maze[newLocation[0]][newLocation[1]] == 2:
            return False
#如果這個位置是正確的話，則回傳True
        else:
            rightRoute.append(newLocation)
            routeHistory.append(newLocation)
            return True

#接下來下面3個Function其實都一樣，只是方向與上方不同(分別是下、左、右)

def goDown(location):
     if(location[0]==6):
        return False
     else:
        newLocation = [location[0]+1, location[1]]
        if newLocation in routeHistory:
            routeHistory.append(newLocation)
            return False       
        elif maze[newLocation[0]][newLocation[1]] == 2:
            return False
        else:
            rightRoute.append(newLocation)
            routeHistory.append(newLocation)
            return True

def goLeft(location):
     if(location[1]==0):
        return False
     else:
        newLocation = [location[0], location[1]-1]
        if newLocation in routeHistory:          
            routeHistory.append(newLocation)
            return False       
        elif maze[newLocation[0]][newLocation[1]] == 2:
            return False
        else:
            rightRoute.append(newLocation)
            routeHistory.append(newLocation)
            return True
        
def goRight(location):
     if(location[1]==6):
        return False
     else:
        newLocation = [location[0], location[1]+1]
        if newLocation in routeHistory:
            routeHistory.append(newLocation)
            return False       
        elif maze[newLocation[0]][newLocation[1]] == 2:
            return False
        else:
            rightRoute.append(newLocation)
            routeHistory.append(newLocation)
            return True
# 設定起始座標
while True:
    x=random.randint(0,6)
    y=random.randint(0,6)
    x1=random.randint(0,6)
    y1=random.randint(0,6)
    if maze[x][y]==0 and maze[x1][y1]==0:
        break

# 這個List用來記錄正確路徑
rightRoute = [[x,y]]
# 記錄已經走過的路徑
routeHistory = [[x,y]]
# 現在的位置
location = [x,y]

# 如果最後一個位置不是在出口[4,4]，則繼續搜尋
while rightRoute[-1] != [x1,y1]:
#以下分別有4個函數，分別代表要向上下左右的方向走，執行函數後，如果回傳True則用下一個位置繼續走，如果回傳False則換方向走
    
    if goRight(location):
        location = rightRoute[-1]
        continue

    if goUp(location):
        location = rightRoute[-1]
        continue
    
    if goDown(location):
        location = rightRoute[-1]
        continue
        
    if goLeft(location):
        location = rightRoute[-1]
        continue
        

#     如果有函數回傳False，代表這個位置沒路走，則把這個位置從正確路徑中剔除
    rightRoute.pop()
#     剔除後重新用新的位置繼續尋找
    location = rightRoute[-1]
# 最後印出路徑
for i in range(len(rightRoute)):
    x=rightRoute[i][0]
    y=rightRoute[i][1]
    maze[x][y]= 3
print(rightRoute)
for i in range(7):
    print(maze[i])
print(routeHistory)
```
